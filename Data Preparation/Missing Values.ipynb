```python
import pandas as pd
import matplotlib.pyplot as plt
import ds_functions as ds
from pandas.plotting import register_matplotlib_converters

register_matplotlib_converters()
heart = pd.read_csv('data/heart_failure_clinical_records_dataset.csv', sep=',', decimal='.', parse_dates=True, infer_datetime_format=True)
```

```python
plt.figure()
mv = {}
for var in heart:
    mv[var] = heart[var].isna().sum()
ds.bar_chart(mv.keys(), mv.values(), title='Nr of missing values per variable',
               xlabel='variables',
               ylabel='nr missing values')
mv.values()
```
